```
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import math
```

```
data =pd.read_csv("fifa_data.csv")
```

```
df2=pd.DataFrame(df['Value'],index=range(1,len(df['Value'])))
```

```
df2
```

```
##ASLI
df3=list(df2['Value'])
p=[]
q=[]
for i in range(0,len(df3)):
  df3[i] = df3[i].replace('€','')
  if df3[i][-1]=='M':
    df3[i] = df3[i].replace('M','')
    p.append(i)
  elif  df3[i][-1]=='K':
    df3[i] = df3[i].replace('K','')
    q.append(i)
  df3[i] = float(df3[i])
df3
```

```
for j in range(0,len(p)):
  df3[p[j]] = df3[p[j]]*1000000
for k in range(0,len(q)):
  df3[q[k]] = df3[q[k]]*1000
df3
```

```
df3.sort()
```

```
print("maximum = ",np.max(df3))
print("minimum = ",np.min(df3))
```

```
a=[]
b=[]
c=[]
d=[]
e=[]
f=[]
for t in range(0,len(df3)):
  if df3[t]<=20000000:
    a.append(df3[t])
  elif df3[t]<=40000000 and df3[t]>20000000:
    b.append(df3[t])
  elif df3[t]<=60000000 and df3[t]>40000000:
    c.append(df3[t])
  elif df3[t]<=80000000 and df3[t]>60000000:
    d.append(df3[t])
  elif df3[t]<=100000000 and df3[t]>80000000:
    e.append(df3[t])
  elif df3[t]<=120000000 and df3[t]>100000000:
    f.append(df3[t])
```

```
a=np.array(a);a1 = math.floor(len(a)/np.max(df3))
b=np.array(b);b1 = math.floor(len(b)/np.max(df3))
c=np.array(c);c1 = math.floor(len(c)/np.max(df3))
d=np.array(d);d1 = math.floor(len(d)/np.max(df3))
e=np.array(e);e1 = math.floor(len(e)/np.max(df3))
f=np.array(f);f1 = math.floor(len(f)/np.max(df3))

```

```
activities = ['0-20M', '20M-40M','40M-60M', '60M-80M', '80M-100M','100M,120M'] 
# portion covered by each label 
slices = [len(a), len(b), len(c), len(d),len(e),len(f)] 
# color for each label 
colors = ['m', 'y', 'g', 'b','r','c'] 
# plotting the pie chart 
plt.pie(slices, labels = activities, colors=colors, 
		startangle=90, shadow = True, explode = (1.5, 1.5, 3.5, 4.5,5.5,6.5), 
		radius = 3.2, autopct = '%5.1f%%')  
# plotting legend 
plt.legend() 
# showing the plot 
plt.show() 
```
